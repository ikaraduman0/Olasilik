###  Bir doa gezisinde, geziye 5 trk, 4 alman, 1 ingiliz, 1 rus, 1 fransz olmak zere toplam 12 kii katlmtr. Yola kan 12 kiilik ekip tek sra halinde yrdne gre grubun en nnde 1 trk en arkasnda ise 1 almann olma olasl nedir? 

# zm

 
 rnek uzaymz = 12! / (5! * 4!) = 166.320 dir.
 
 Olay {T...A,...} 
 1 Trk nde, 1 Alman arkada olmak zere,
 
 Kalan Durum = 10! / (4! * 3!) = 25.200 dr.
  
 Olaslk deeri = 25200 / 166320 = 0,1515 = % 15,15
 

```python
import math  

def Faktoriyel(sayi):
    fakt=1
    for i in (range(1,sayi+1)):
        fakt=fakt*i
    
    return (fakt)

Ornekuzay=Faktoriyel (12) / (Faktoriyel(5)*Faktoriyel(4))
KalanDurum=Faktoriyel (10) / (Faktoriyel(4)*Faktoriyel(3))
Olasilik = KalanDurum / Ornekuzay

print ("rnek Uzay : ", Ornekuzay)
print ("Kalan Durum: ", KalanDurum)
print ("Olaslk % : ", Olasilik*100)
```
