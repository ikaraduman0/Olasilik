###  Bir doğa gezisinde, geziye 5 türk, 4 alman, 1 ingiliz, 1 rus, 1 fransız olmak üzere toplam 12 kişi katılmıştır. Yola çıkan 12 kişilik ekip tek sıra halinde yürüdüğüne göre grubun en önünde 1 türk en arkasında ise 1 almanın olma olasılığı nedir? 

# Çözüm

 
 Örnek uzayımız = 12! / (5! * 4!) = 166.320 dir.
 
 Olay {T...A,...} 
 1 Türk önde, 1 Alman arkada olmak üzere,
 
 Kalan Durum = 10! / (4! * 3!) = 25.200 dür.
  
 Olasılık değeri = 25200 / 166320 = 0,1515 = % 15,15
 

```python
import math  

def Faktoriyel(sayi):
    fakt=1
    for i in (range(1,sayi+1)):
        fakt=fakt*i
    
    return (fakt)

Ornekuzay=Faktoriyel (12) / (Faktoriyel(5)*Faktoriyel(4))
KalanDurum=Faktoriyel (10) / (Faktoriyel(4)*Faktoriyel(3))
Olasilik = KalanDurum / Ornekuzay

print ("Örnek Uzay : ", Ornekuzay)
print ("Kalan Durum: ", KalanDurum)
print ("Olasılık % : ", Olasilik*100)
```
